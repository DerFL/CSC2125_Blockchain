# Block Chain Slide 1

## What is Blockchain

Blockchain is a list of ordered data blocks maintained by a network that

- Every one can read
- Every one can append a valid block at the end
- No one can modify existing blocks



## Cryptocurrecy is Blockchain Stroing Transactions

- The whole transaction history is maintained by a network


## The Promise of Blockchain & Cryptocurrency
A system that is
- Decentralized (No central point of failure)
- Trustless (No central controls)
- Anonymous (Difficult to track real-world identity)

- Wallet creation generates a Bitcoin address

- Share address to receive money
- Sender specifies address and amount
- Transaction is broadcast to the network, "miners" verify it and add to the transaction history
