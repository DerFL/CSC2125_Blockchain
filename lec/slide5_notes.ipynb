# Block Chain Slide 5

### Standard block chain have very limited throughput

- Bitcoin: 1MB block per 10 minutes
- Each transaction is at least 250-300B
- Slow/small block generation



## Attempts

- SegWit (Reduce transaction size)
- Increase block size
- Faster block generation
  - Introduce many forks (wasteful)
  - Downgrade safety (easier to beat longest chain)
- Maually elected nodes
- Rotating committee
- Sharding

## Proof-of-Stake

- PoW is energy inefficient (work done is not useful)
- Need a better way to define voting power
- PoS is an anti-sybil-attack mechanisim, designed to replace PoW.
- It is not targeted to improve Blockchain throughput


### Problems and Challenges with PoS


- Nothing At Stake
  - Incentivize generating blocks on every block
  - Penalties designed to discourage this behavior

- Long range attack
